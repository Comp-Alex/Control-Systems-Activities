```python
import control as ctrl
import matplotlib.pyplot as plt
```

```python
num = [5, 5]
den = [1, 4, 3]
G = ctrl.tf(num, den)
print(G)
```

```python
poles = ctrl.poles(G)
print("Poles:", poles)
```

```python
zeros = ctrl.zeros(G)
print("zeros:", zeros)
```

```python
ctrl.pzmap(G, grid=True)
plt.show()
```
